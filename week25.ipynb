###Frequency of selected categorical variables with histogram/bar chart

```python
charges_smoker_avg = df.groupby(['non_smoker_vehicle'])['colour'].mean()
```

```python
fig = px.bar(charges_smoker_avg, x=charges_smoker_avg.index, y='colour',
            labels={'x':'non_smoker_vehicle'})

fig.update_layout(title="Average colour by Smoking")

fig.show()
```

```python
import plotly as py
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

###Frequency of selected categorical variables with pie chart/donut

```python
fig = make_subplots(rows=2, cols=2, specs=[[{"type": "pie"}, {"type": "pie"}], 
                                           [{"type": "pie"}, {"type": "pie"}]],
                    subplot_titles=("make", 
                                    "gearbox",
                                    "colour",
                                    "upholstery_colour"))

fig.add_trace(go.Pie(labels=data1.make.value_counts().index, 
                    values=data1.make.value_counts().values,
                     name="make"),
             row=1, col=1)

fig.add_trace(go.Pie(labels=data1.gearbox.value_counts().index, 
                    values=data1.gearbox.value_counts().values,
                     name="gearbox"),
             row=1, col=2)

fig.add_trace(go.Pie(labels=data1.colour.value_counts().index, 
                    values=data1.colour.value_counts().values,
                     name="colour"),
             row=2, col=1)

fig.add_trace(go.Pie(labels=data1.upholstery_colour.value_counts().index, 
                    values=data1.upholstery_colour.value_counts().values,
                     name="upholstery_colour"),
             row=2, col=2)

fig.update_layout(showlegend=False, title="Frequency of Categorical variables",
                 height = 700, width = 900)

fig.show()
```
