## General Information about data set

**Description of the variables:**

**Price:** car price in euro -----numerical variables---

**Make:** car brand -----categorical variables---

**Mileage:** total km    -----numerical variables---

**Fuel_Type:** type of fuel used in car   -----categorical variables---

**Power:** car power     -----numerical variables---

**Seller:** Real car Agent  -----categorical variables---

**body_type:** car body type  ----categorical variables---

**seats :** number of car's seat ----categorical variables---

**non_smoker_vehicle:** smoke usage  ----categorical variables---

**engine_size:** car's engine size  -----numerical variables---

**gears:** type of car's gears  ----categorical variables---

**co2_emissions:** emission release -----numerical variables---

**avg_mileage:** average mileage -----numerical variables---

**age:** car age  -----numerical variables---

```python
make = df.groupby(['make']).count()
fig = px.histogram(df, x="make", color='body_type')
fig.show()
```

```python
price_make_avg = df.groupby(['make'])['price'].mean()
price_make_sum = df.groupby(['make'])['price'].sum()
```

```python
fig = px.bar(price_make_sum, x=price_make_sum.index, y='price',
            labels={'x':'model'})

fig.update_layout(title="Total Prices by Make")

fig.show()
```

## Heatmap for correlation between numerical variables**

```python
continuous_features  = ['price','mileage','power','cylinders','avg_mileage','age','comb']
fig = px.imshow(df[continuous_features].corr());
fig.show()
```

### Heatmap
- Price – power
- Mileage – age
- power – comb
- cylinder - avg_mileage
- age – comb
